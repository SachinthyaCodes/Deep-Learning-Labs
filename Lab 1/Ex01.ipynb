```python
# 01.1

import numpy as np          # we have to import the numPy first

# create 4x4 array exponential distribution
arr4x4 = np.random.exponential(1,(4,4))       # 1st parameter -> scale and second parameter size of array
print(arr4x4)
```

```python
# 01.2


import matplotlib.pyplot as plt     # we have to import matplotLib to get statics and animated data vizualizations
# pyplot --> generte some charts

arr1 = np.random.exponential(1,100000)
uniformDist = np.random.rand(100000)
normalDist = np.random.rand(100000)

# create the visualization
plt.hist(arr1, density=True, bins=100, histtype="step", color="red", label="Exponential")
plt.hist(uniformDist, density=True, bins=100, histtype="step", color="green", label="Uniform")
plt.hist(normalDist, density=True, bins=100, histtype="step", color="blue", label="Normal")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
#01.3


from mpl_toolkits.mplot3d import Axes3D   # provide 3D plottings


x = np.linspace(-5, 5, 200)   # x in grid with mentioned range
y = np.linspace(-5, 5, 200)   # y -----------
X, Y = np.meshgrid(x, y)  # cerate meshgrid
Z = X**2 + Y**2

# Plotting the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title('3D Plot of Z = X^2 + Y^2')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

```python
#01.4

import seaborn as sns    # visulaze some data in python
import pandas as pd

data = np.random.rand(100,6)
col = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
df = pd.DataFrame(data , columns=col)

# Calculate correlation
correlation_matrix_pearson = df.corr(method='pearson')
correlation_matrix_spearman = df.corr(method='spearman')


plt.figure(figsize=(10, 8))   # heatmap for pearson
sns.heatmap(correlation_matrix_pearson, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Pearson Correlation Heatmap')
plt.show()


plt.figure(figsize=(10, 8))  # heatmap for spearman
sns.heatmap(correlation_matrix_spearman, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Spearman Correlation Heatmap')
plt.show()
```
