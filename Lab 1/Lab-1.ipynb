#Note: Differnece Between Uniform , Normal and Exponential distribution

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate sample data
uniform_data = np.random.uniform(low=0, high=10, size=10000)      # equally likely values between 0 and 10
normal_data = np.random.normal(loc=5, scale=1.5, size=10000)      # mean=5, std deviation=1.5
exponential_data = np.random.exponential(scale=2.0, size=10000)   # mean=2.0

# 2. Plot histograms
plt.figure(figsize=(12, 4))

# Uniform
plt.subplot(1, 3, 1)
plt.hist(uniform_data, bins=30, color='skyblue', edgecolor='black')
plt.title("Uniform Distribution\nEqual chance for all values")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Normal
plt.subplot(1, 3, 2)
plt.hist(normal_data, bins=30, color='lightgreen', edgecolor='black')
plt.title("Normal Distribution\nBell curve around mean")
plt.xlabel("Value")

# Exponential
plt.subplot(1, 3, 3)
plt.hist(exponential_data, bins=30, color='salmon', edgecolor='black')
plt.title("Exponential Distribution\nMost values near 0")
plt.xlabel("Value")

plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt

# Generate random data samples from different distributions
exp_data = np.random.exponential(scale=1, size=(10000,))  # Exponential distribution with mean=1
uniform_data = np.random.rand(100000)                     # Uniform distribution between 0 and 1
normal_data = np.random.randn(100000)                     # Standard normal distribution (mean=0, std=1)

# Create a new figure with specified size
plt.figure(figsize=(10,6))

# Plot histogram for exponential data
plt.hist(exp_data, density=True, bins=200, histtype="step", color="red",
         label="Exponential", alpha=0.8)

# Plot histogram for uniform data
plt.hist(uniform_data, density=True, bins=200, histtype="step", color="blue",
         label="Uniform", alpha=0.8)

# Plot histogram for normal data
plt.hist(normal_data, density=True, bins=200, histtype="step", color="green",
         label="Normal", alpha=0.8)

# Set axis limits: x-axis from -2.5 to 8, y-axis from 0 to 1.1
plt.axis([-2.5, 8, 0, 1.1])

# Add legend in the upper left corner to identify distributions
plt.legend(loc="upper left")

# Add title and axis labels
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")

# Add light gridlines for better readability
plt.grid(alpha=0.3)

# Display the plot
plt.show()
```

```python
# Task 1.3 - 3D plot for Z = X^2 + Y^2
from mpl_toolkits.mplot3d import Axes3D

# Create grid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

# Plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap="plasma", edgecolor="none")

ax.set_title("3D Surface: Z = XÂ² + YÂ²")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")

plt.show()
```

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Generate random data (100 samples, 6 features)
data = np.random.rand(100, 6)
col = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
df = pd.DataFrame(data, columns=col)

# Calculate correlation
correlation_matrix_pearson = df.corr(method='pearson')
correlation_matrix_spearman = df.corr(method='spearman')

# Pearson Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_pearson, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Pearson Correlation Heatmap')
plt.tight_layout()
plt.show()

# Spearman Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_spearman, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Spearman Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("./Au_nanoparticle_dataset.csv")

new_df =  df[['N_total', 'N_bulk', 'N_surface', 'R_avg']] # new data frame with selected columns
```

```python
new_df.head(5)
```

```python
mean_values = new_df.mean()
st_d = new_df.std()
quartiles =  new_df.quantile([0.25, 0.5, 0.75])    # define q1 , q2 ,q3

print("Mean Values" ,mean_values)
print("Std values : ",st_d)
print("Quartiles :",quartiles)
```

```python

plt.figure(figsize=(16, 4))

for i, column in enumerate(new_df.columns, 1):
    plt.subplot(1, 4, i)
    plt.hist(new_df[column], bins=30, color='skyblue', edgecolor='black')
    plt.title(f'{column} Histogram')
    plt.xlabel(column)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
sns.pairplot(new_df)
plt.show()
```
